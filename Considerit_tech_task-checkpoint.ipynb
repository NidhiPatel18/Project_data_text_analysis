<h3>Keyword Extraction using TF-IDF  </h3>

TF-IDF (term frequency-inverse document frequency) is a weighting statistic that indicates if a word is important in a particular document of a corpus. 
This is done by multiplying two metrics:
1) TF: how many times a word appears in a document,
2) IDF: the inverse document frequency of the word across a set of documents.This means, how common or rare a word is in the entire document set.

Performing pre-processing for the feature extarction on the objective column of project to get important topics from each project.

```python
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
import re
from nltk.stem.snowball import SnowballStemmer
from nltk.stem import WordNetLemmatizer
 
nltk.download('wordnet')
nltk.download('averaged_perceptron_tagger')
nltk.download('stopwords')
nltk.download('omw-1.4')
stop_words = set(stopwords.words('english'))


def preprocess(input_text,flg_stem,flg_lem):

#removing special charachters
    processed_text = re.sub("[^a-zA-Z]", " ", input_text)
    
#tokenizing the sentence and coverting into lowercase
    token_words = processed_text.lower().split()
    final_tokens=[]
    
#removing stop words
    for word in token_words:
        if word not in stop_words:
            final_tokens.append(word)
        
#Applying stemming (here we are using the snowballstemmer as it is improvised version of the porterstemmer, lancasterstemmer tends to over stem the data and hence we can loose meaning of the word) )
    if flg_stem == True:
        s_stemmer = SnowballStemmer(language='english')
        result_tokens=[]
        for word in final_tokens:
            word = s_stemmer.stem(word)  
            result_tokens.append(word)

#Lemmatize verbs in list of tokenized words
    if flg_lem == True:
        lemmatizer = nltk.stem.wordnet.WordNetLemmatizer()
        result_tokens=[]
        for word in final_tokens:
            #tagged= nltk.pos_tag(word)
            word = lemmatizer.lemmatize(word)
            result_tokens.append(word)
            
#returning the processed string
    return " ".join(result_tokens)
            
        
```

Stemming follows an algorithm with steps to perform on the words which makes it faster. Whereas, in lemmatization uses a corpus also to supply lemma which makes it slower than stemming. Therefore, if speed is concern then stemming should be used since lemmatizers scan a corpus which consumes time and processing on the other hand if we wnt accuracy, lemmatization is advisable. Here, both are used in order to see a difference between both.

```python
test_sentence = "this is a sentence to demonstrate how the preprocessing function preprocesses... 1234!"

print("with the stamming: \n",preprocess(test_sentence,True,False))
```

```python
test_sentence1="Player playing a play on a playground!"

print("with the lemmatizer: \n",preprocess(test_sentence1,False,True))
```

Implimenting TF-Idf aldorithm for keyword extraction.

```python

from sklearn.feature_extraction.text import TfidfVectorizer
import math
from collections import Counter

def tf_idf_extractor(data):
    tfidf = TfidfVectorizer()
    tfs = tfidf.fit_transform([data]) 

    feature_names = tfidf.get_feature_names()

    df = pd.DataFrame(tfs.T.toarray(), index=feature_names, columns= ['weight'])
    df = df.sort_values(by = 'weight', ascending = False)
    word_lists = df.index.values
    unique_str  = ' '.join(word_lists)
    return unique_str


```

<h3>Applying RAKE algoridthm for keyword extraction </h3>

RAKE short for Rapid Automatic Keyword Extraction algorithm, is a domain independent keyword extraction algorithm which tries to determine key phrases in a body of text by analyzing the frequency of word appearance and its co-occurance with other words in the text.This powerful keyword extraction method is performed by combining the RAKE algorithm with the NLTK toolkit. It is known as rake-nltk. Here, get_ranked_phrases method returns 10 ranked keywords.

```python
from rake_nltk import Rake
nltk.download('punkt')

def rake_extractor(text):
    
    r = Rake()
    r.extract_keywords_from_text(text)
    return r.get_ranked_phrases()[:10]
```

```python
project_1['keywords_rake']= project_1['desc'].apply(rake_extractor)
```

<h3>Extracting keywords using POS- Tagging</h3>

The main idea behind POS-tagging is to assign special label  to each token (word) in a text corpus to indicate the part of speech.There are eight main Parts of Speech: Nouns(naming word), Pronouns(replaces a noun), Adjectives(describing word), Verbs(action word), Adverbs(describes a verb), Prepositions(shows relationships), Conjunctions(joining word) and Interjections(Expressive word). Each part can be used for different conserns for example, in order to perform sentimantal analysis, we would focus on Adverbs and Adjectives. However, here our concern is to to extract the main topics of the text, we are considering Nouns. The task has been fylfilled by spaCy,  a free and open-source library for Natural Language Processing (NLP) in Python with a lot of in-built capabilities. 

```python
# import spacy and load english language model
import spacy
nlp = spacy.load('en_core_web_sm',disable=['ner','textcat'])


def tag_model(text):
    
# create spacy object
    doc = nlp(text)
    
    keywords=[] 
    for token in doc:
    # check token pos
        if token.pos_=='NOUN':        # print token
            keywords.append(token)
    
    return keywords
```

```python
project_1['keywords_pos']= project_1['desc'][0:500,].apply(tag_model)
```

<h3> Name Entity Reccognization </h3>

Now, let's perform Name Entity Reccognization. The basic idea behind it is to locate and classify named entites into text into predefined ctagories such as person, locations organizations, amount, percentages. To answer key questions about text. One simple way to implement it is with open source library called spaCy. Entities has been color-coded further with the help of displacy library in order to get better intuation.

```python
from spacy import displacy
NER= spacy.load('en_core_web_sm')
def name_entity(text):
    doc= NER(text)
    displacy.render(doc, style='ent', jupyter=True)
    
```

<h3> Applying k-means clustering to classifying and pridicting the project domain</h3>

In this section we will apply k-means clustering, an unsupervised learning algorithm to classifiy the document into k clusters. In order to achieve more accurate results, we have given the list of priviously extracted keywords(keywords_rake) into TfidfVectorizer as an input rather than raw text. TfidfVectorizer converts the list of keywords into array of numbers and K-means algorithm further classifies them in diffrent clusters by mapping the distance between each of them.

```python
#converting the list of keyword in the format accepted by usingTfidfVectorizer.

textlist = []
textlist = list(project_1['keywords_rake'].iloc[:500, ])
flat_list = []
for sublist in textlist:
    for item in sublist:
        flat_list.append(item)
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.cluster import KMeans
from sklearn import metrics
import math

#create vectorizer usingTfidfVectorizer class to fit and transform the document
vectorizer = TfidfVectorizer(stop_words='english')
X = vectorizer.fit_transform(flat_list).toarray()


#implementing k-means clustering algorithm in vectorized document
true_k = 20
model = KMeans(n_clusters=true_k, init='k-means++', max_iter=100, n_init=1)
model.fit(X)

#getting the centroids and features
order_centroids = model.cluster_centers_.argsort()[:, ::-1]
terms = vectorizer.get_feature_names()

# print the centroids into which clusters they belongs
for i in range(true_k):
    print('\n'+'Cluster:'+ str(i)+'\n'),
    for ind in order_centroids[i, :10]:
        print(terms[ind])
```

FInally,predicting the project title which can tell us in which cluster the given project belongs to.

```python
pred_str="Improving digital health literacy in Europe'"

print('\n')
print('Prediction:')
X = vectorizer.transform([pred_str])
predicted = model.predict(X)
print(predicted)
```
